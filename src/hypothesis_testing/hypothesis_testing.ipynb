```python
#########################################
### group Emotion vs personality  ####
#########################################


#########################################
### group Enjoyment vs personality  ####
#########################################
personalityS = data_g.loc[data.Personality=='S']
personalityC = data_g.loc[data.Personality=='C']
personalityI = data_g.loc[data.Personality=='I']
personalityD = data_g.loc[data.Personality=='D']



result = stats.kruskal(personalityS.Enjoyment.tolist(),personalityC.Enjoyment.tolist(),personalityI.Enjoyment.tolist(),personalityD.Enjoyment.tolist())
 
# Print the result
print(result)
print('personality influences enjoyment')

#########################################
### group boredom vs personality  #####
#########################################

result = stats.kruskal(personalityS.Boredom.tolist(),personalityC.Boredom.tolist(),personalityI.Boredom.tolist(),personalityD.Boredom.tolist())
 
# Print the result
print(result)
print('personality influences Boredom ')

#########################################
### group Confidence vs personality  ####
#########################################

result = stats.kruskal(personalityS.Confidence.tolist(),personalityC.Confidence.tolist(),personalityI.Confidence.tolist(),personalityD.Confidence.tolist())
 
# Print the result
print(result)
print('personality influences Confidence')

#########################################
### group Anxiety vs personality  ####
#########################################

result = stats.kruskal(personalityS.Anxiety.tolist(),personalityC.Anxiety.tolist(),personalityI.Anxiety.tolist(),personalityD.Anxiety.tolist())
 
# Print the result
print(result)
print('personality influences Anxiety')
```
